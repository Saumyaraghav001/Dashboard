```python
crosstab = pd.crosstab(df['gender'], df['Churn'], normalize='index') * 100  # Percent per gender

# Step 2: Plot
ax = crosstab.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='Set1')

# Step 3: Add % labels on each segment
for bars in ax.containers:
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{height:.1f}%',
                ha='center',
                va='center',
                fontsize=10,
                color='black'
            )

# Step 4: Formatting
plt.title('Churn % by Gender (Stacked)')
plt.ylabel('Percentage')
plt.xlabel('Gender')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
services = ['PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid: 3 rows x 3 columns
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Countplots for Service Features', fontsize=16)

# Flatten axes array for easy indexing
axes = axes.flatten()

# Loop through services and create countplots
for i, col in enumerate(services):
    sb.countplot(data=df, x=col, ax=axes[i], hue=df'Churn',palette='Set2')
    axes[i].set_title(f'{col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

# Hide any unused subplots (in case list length < grid size)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0, 1, 0.97])  # Adjust layout to fit the suptitle
plt.show()
```

#The visualized data shows the relationship between customer churn and various telco services.
Customers who do not use additional services like OnlineSecurity, TechSupport, and DeviceProtection tend to churn more. 
On the other hand, services like PhoneService and MultipleLines show less obvious differences in churn behavior.
Notably, customers with Fiber optic Internet have a significantly higher churn rate compared to DSL users.
